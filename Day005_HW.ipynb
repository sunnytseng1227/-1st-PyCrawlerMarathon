```python
#請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：
#1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
#2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
#3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
```

```python
import requests,json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
getlist_ = json.loads(r.text)
print("這個 API 一次會回傳幾筆資料？",len(getlist_),"筆\n每一筆資料包含哪些欄位？")
print(getlist_[0].keys())
```

```python
def avg(ns):
    sum=0
    for n in ns:
        sum=sum+n
    #print(ns,"的平均值 = ",sum/len(ns)) #加總後除於長度
    return sum/len(ns)
```

```python
import requests,json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
getlist_ = json.loads(r.text)
_commentCount, _likeCount= [], []

for rows in getlist_:
    _commentCount.append(rows['commentCount'])
    _likeCount.append(rows['likeCount'])
    print("貼文時間:{0:30s}留言人數/按讚人數:{2:4d}/{3:4d}   標題:{1}".format(rows["createdAt"],rows["title"],rows["commentCount"],rows["likeCount"]))

print("\n\n熱門文章各則留言人數：",_commentCount)
print("熱門文章平均留言人數:",avg(_commentCount))
print("熱門文章各則按讚人數：",_likeCount)
print("熱門文章平均按讚人數:",avg(_likeCount))
```

```python
import requests,json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
getlist_ = json.loads(r.text)
_commentCount, _likeCount= [], []

for rows in getlist_:
    _commentCount.append(rows['commentCount'])
    _likeCount.append(rows['likeCount'])
    print("貼文時間:{0:30s}留言人數/按讚人數:{2:4d}/{3:4d}   標題:{1}".format(rows["createdAt"],rows["title"],rows["commentCount"],rows["likeCount"]))

print("\n\n非熱門文章各則留言人數：",_commentCount)
print("非熱門文章平均留言人數:",avg(_commentCount))
print("非熱門文章各則按讚人數：",_likeCount)
print("非熱門文章平均按讚人數:",avg(_likeCount))
```
